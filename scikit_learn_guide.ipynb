## 붓꽃(Iris) 데이터 품종 예측하기

```python
#데이터 로드 라이브러리
from sklearn.datasets import load_iris
#모델 업로드
from sklearn.tree import DecisionTreeClassifier
#학습데이터 테스트데이터 분리코드
from sklearn.model_selection import train_test_split
import pandas as pd

# 붓꽃 데이터 세트를 로딩합니다.
iris = load_iris()

# iris.data 변수에 저장.
# iris.data는 Iris 데이터 세트에서 피처(feature)만으로 된 데이터를 numpy로 가지고 있습니다.
iris_data = iris.data

# iris.target 변수에 저장.
# iris.target은 붓꽃 데이터 세트에서 레이블(결정 값) 데이터를 numpy로 가지고 있습니다.
iris_label = iris.target
print('iris target값:', iris_label)

# iris target명 확인 0='setosa', 1='versicolor', 2='virginica'
print('iris target명:', iris.target_names)

# 붓꽃 데이터 세트를 자세히 보기 위해 DataFrame으로 변환합니다.
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)

# 모델 학습 시작
dt_clf.fit(iris_data, iris_label)

# 학습 데이터 셋으로 예측 수행
pred = dt_clf.predict(iris_data)

# 예측 정확도를 구하기 위한 accuracy_score import
from sklearn.metrics import accuracy_score
print('예측 정확도:',accuracy_score(iris_label, pred))
```

### train_test_split
전체 데이터를 tarin, test 데이터로 분리하는 과정이 필요하다.

```python
# data, label을 train, test 데이터로 분리
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)
학습용_data, 테스트용_data,학습용_target, 테스트용_target = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)

# 학습 수행
dt_clf.fit(X_train, y_train)

# 학습이 완료된 DecisionTreeClassifier 객체에서 테스트 데이터 세트로 예측 수행.
pred = dt_clf.predict(X_test)

# 에측해보기
from sklearn.metrics import accuracy_score
print("예측 정확도: {0:.4f}".format(accuracy_score(y_test, pred)))
```
