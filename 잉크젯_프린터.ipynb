```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('InkjetDB_preprocessing.csv')

#data, target 분리
X = df.drop(['PatternSize'], axis=1)
y = df['PatternSize']
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 1)

print("학습 데이터:",X_train.shape)
print("테스트 데이터:", X_test.shape)
print("학습 데이터:",y_train.shape)
print("테스트 데이터:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("결정 계수:", r2)
print("평균 제곱근 오차:", rmse)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error

#회귀 모델 정의
KN = KNeighborsRegressor()
#모델 학습
KN.fit(X_train, y_train)
#예측
y_pred = KN.predict(X_test)
#결정 계수 계산
r2 = r2_score(y_test, y_pred)
#평균 제곱근 오차 계산
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(r2)
print(rmse)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error

svm_regressor = SVR()

svm_regressor.fit(X_train, y_train)

y_pred = svm_regressor.predict(X_test)

r2 = r2_score(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(r2)
print(rmse)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score, mean_squared_error

ens = AdaBoostRegressor()

ens.fit(X_train, y_train)

y_pred = ens.predict(X_test)

r2 = r2_score(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(r2)
print(rmse)
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import r2_score, mean_squared_error

ense = BaggingRegressor()
ense.fit(X_train,y_train)

y_pred = ense.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(r2)
print(rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

model = RandomForestRegressor()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

r2 = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)

print(r2)
print(rmse)
```
